# Using List Comprehension structure, capitalise the names of numeric variables in car_crashes data and add NUM at the beginning

```python
import seaborn as sns
df=sns.load_dataset("car_crashes")
df.columns
```

```python
df.columns=[col.upper() for col in df.columns]
["NUM_"+ col if df[col].dtype != "O" else col for col in df.columns ]
```

# Using List Comprehension structure, write "FLAG" at the end of the names of variables that do not contain "no" in their names in car_crashes data

```python
df=sns.load_dataset("car_crashes")
df.columns

[col.upper() + "_FLAG" if "no" not in col else col.upper() for col in df.columns]
```

# Using List Comprehension structure, select the names of the variables that are different from the variable names given below and create a new dataframe.First create a new list named new_cols using list comprehension according to the given list. Then create a new df by selecting these variables with df[new_cols] and name it new_df

```python
import seaborn as sns
import pandas as pd

pd.set_option("display.max_columns",None)
pd.set_option("display.float_format",lambda x: "%.3f" % x)
pd.set_option("display.width",500)
```

```python
df=sns.load_dataset("car_crashes")
df.columns
df.head()

og_list = ["abbrev","no_previous"]
```

```python
new_cols=[col for col in df.columns if col not in og_list]

new_df=df[new_cols]
new_df.head()
```
